```python
def part_1(lines):
    pairs = list(map(lambda x: x.strip().split(' '), lines))
    score = 0
    for pair in pairs:
        pick1 = ord(pair[0]) - ord('A')
        pick2 = ord(pair[1]) - ord('X')
        score += pick2 + 1
        if pick1 == pick2:
            score += 3
        elif (pick1 + 1) % 3 == pick2:
            score += 6
    return score
```

```python
def part_2(lines):
    pairs = list(map(lambda x: x.strip().split(' '), lines))
    score = 0
    for pair in pairs:
        pick1 = ord(pair[0]) - ord('A')
        if pair[1] == 'Y': # DRAW
            score += 3 + pick1 + 1
        elif pair[1] == 'X': # LOSE
            score += (pick1 + 3 - 1) % 3 + 1
        elif pair[1] == 'Z': # WIN
            score += 6 + (pick1 + 1) % 3 + 1
        else:
            raise Exception("Invalid input")
    return score
```

```python
with open("../inputs/day02.txt", 'r') as f:
    lines = f.readlines()
    print("Part 1:", part_1(lines))
    print("Part 2:", part_2(lines))
```
